```
df_copy=df.copy()
name_year = df_copy[df_copy.duplicated('Name',keep = False) & (df_copy['Year'].isnull())]['Name'].unique()
df_copy[df_copy['Name'].isin(name_year)].sort_values('Name')
```

```
publisher_year = df_copy[df_copy.duplicated('Name',keep = False) & (df_copy['Publisher'].isnull())]['Name'].unique()
df_copy[df_copy['Name'].isin(publisher_year)].sort_values('Name')
```

```
genre_year = df_copy[df_copy.duplicated('Name',keep = False) & (df_copy['Genre'].isnull())]['Name'].unique()
df_copy[df_copy['Name'].isin(genre_year)].sort_values('Name')
```

```
data_columns = ['Year','Publisher','Genre']

for col in data_columns:
  #중복된 Name에 'Year','Publisher','Genre'컬럼이 nan인 경우
  dn = df_copy[df_copy.duplicated('Name') & df_copy[col].isnull()]

  #같은 Name의 'Year','Publisher','Genre' 값을 가져와서 대체하기
  for name in dn['Name'].unique():
    value = df_copy.loc[df_copy['Name']==name,col].dropna().sort_values(ascending = False)
    value_fvi = value.first_valid_index()
    if value_fvi is not None:
      result = df_copy.loc[value_fvi, col]
    df_copy.loc[(df_copy['Name'] == name) & df_copy[col].isnull(), col] = result
     
```

```

df_copy.isnull().sum().to_frame() #결측치 확인
```

```
df_copy.dropna(subset=['Year','Genre','Publisher'], inplace=True)
df_copy.isnull().sum().to_frame()
```

```
df=df_copy.copy()
```

```
df=df.astype({'Year':'int'})  #year 컬럼 년도를 int 타입으로 변경경
df.head()
```

#### 판매량 컬럼의 K랑 M 이상치 EDA

```
df_tOth=df_t[df_t['Other_Sales'].str.contains('K')].copy()
df_tOth

import re

def cleaning_Oth(df_test):
  li_Oth=[]
  for row in df_test['Other_Sales']:
    result = re.sub(r"[^0-9]",'', row)  # .기준 앞 부분만 활용하여 숫자가 아닌 문자는 공백으로 치환합니다.
    li_Oth.append(int(result)/1000)
  df_test['Other_Sales']=li_Oth
cleaning_Oth(df_tOth)
df_tOth

df_tnew=df_tOth.combine_first(df_t).copy()
df_tnew

df_OthM=df_tnew[df_tnew['Other_Sales'].str.contains('M',na=False)].copy()
df_OthM

df_OthM=df_tnew[df_tnew['Other_Sales'].str.contains('M',na=False)].copy()

def cleaning_Oth_M(df_test):
  li_Oth=[]
  for row in df_test['Other_Sales']:
    result = re.sub(r"[^0-9]",'', row)  # .기준 앞 부분만 활용하여 숫자가 아닌 문자는 공백으로 치환합니다.
    li_Oth.append(float(result)/100)
  df_test['Other_Sales']=li_Oth
cleaning_Oth_M(df_OthM)
df_OthM

df_fin_Oth=df_OthM.combine_first(df_tnew).copy()
df_fin_Oth
```

```
# StandardScaler()를 사용하여 데이터 표준화 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
scaler = StandardScaler()
Z = scaler.fit_transform(df_Gen_year2)
print("\n Standardized Data: \n", Z)

# 표준화한 데이터에 대하여 pca 시행 
pca = PCA() 
pca.fit(Z)

print("\n Eigenvectors: \n", pca.components_)
print("\n Eigenvalues: \n",pca.explained_variance_)

B = pca.transform(Z)
print("\n Projected Data: \n", B)
```

```
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

sse = {}
# 최종 모델 학습
for k in range(1,10):
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(df_Gen_year2)
  sse[k] = kmeans.inertia_
  
# 결과 시각화

plt.title('The Elbow Method')
plt.xlabel('Values of k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```
from sklearn.metrics import silhouette_score
best_silhouette_score=-1
for k in range(1,10):

  kmeans = KMeans(n_clusters = 6).fit(df_Gen_year2)
  labels = kmeans.labels_
  score=silhouette_score(df_Gen_year2, labels, metric = 'euclidean')
  print('k:', k,'score:',score)
  if score>best_silhouette_score :
      best_n=k
      best_silhouette_score=score

print(best_n,best_silhouette_score)
```
